```python
from elasticsearch import Elasticsearch
from elasticsearch.helpers import bulk
from elasticsearch.exceptions import ConnectionError
import pandas as pd
import time
from tweepy.streaming import Stream
from tweepy import OAuthHandler
from tweepy import Stream
from textblob import TextBlob
import json

# follow Mihai Todor's suggestion on https://stackoverflow.com/questions/48711455/create-dockerized-elasticsearch-index-using-a-python-script-running-in-docker/48712414#48712414
es = Elasticsearch(hosts=[{"host": "elasticsearch"}], retry_on_timeout=True)

for _ in range(100):
    try:
        # make sure the cluster is available
        es.cluster.health(wait_for_status="yellow")
    except ConnectionError:
        time.sleep(2)
```

```python
class TweetStreamListener(Stream):

    # on success
    def on_data(self, data):
        try:
            dict_data = json.loads(data)


            tweet = TextBlob(dict_data["text"])


#             print(tweet.sentiment.polarity)


            if tweet.sentiment.polarity < 0:
                sentiment = "negative"
            elif tweet.sentiment.polarity == 0:
                sentiment = "neutral"
            else:
                sentiment = "positive"



            # add text and sentiment info to elasticsearch
            try:
                es.index(index="covid19",
                         doc_type="test-type",
                         document={"user": dict_data["user"]["screen_name"],
                                   "time": dict_data["created_at"],
                                   "polarity": tweet.sentiment.polarity,
                                   "subjectivity": tweet.sentiment.subjectivity,
                                   "sentiment": sentiment})
#                 print(tweet)
#                 print(sentiment)
            except:
                print("-------------------error-------------------")
        except:
            print("-------------------error2-------------------")
            
        return True
```
